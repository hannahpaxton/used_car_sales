The max and min values do not make sense in the context of years and would seem to be inaccurate

```python
reg_year_sorted_high = autos["registration_year"].value_counts().sort_index(ascending=False).head(50)
reg_year_sorted_low = autos["registration_year"].value_counts().sort_index().head(50)

reg_year_sorted_high
```

```python
reg_year_sorted_low
```

The web crawling exercise stopped in 2016, therefore, registration values post-2016 don't make sense here. When it comes to minimum values, there is a distict jump from 1800 - 1910 (with 58 records). It makes sense to me to set 2016 as the upper bound and 1910 as the lower bound (inclusive). 

```python
(~autos["registration_year"].between(1910,2016,inclusive="both")).sum() / autos.shape[0]
```

Less than 4% of data, I will move forward with removing these outliers 

```python
autos = autos[autos["registration_year"].between(1910,2016,inclusive="both")]
autos["registration_year"].value_counts(normalize=True)
```

```python
brand_counts = autos["brand"].value_counts(normalize=True)

brand_counts
```

```python
aggregation_brands = brand_counts[brand_counts > .05].index

aggregation_brands
```

```python
brand_mean_price = {}

for brand in aggregation_brands:

    autos_brand = autos[autos["brand"] == brand]
    mean_price = autos_brand["price"].mean()
    brand_mean_price[brand] = mean_price

brand_mean_price
```

These brands come in at different price points categories - while Audi, BMW, and Mercedez Benz are high, Ford & Opel are low. Volkswagen sits between these categories.

```python
brand_mean_km = {}

for brand in aggregation_brands:

    autos_brands = autos[autos["brand"] == brand]
    mean_km = autos_brands["odometer_km"].mean()
    brand_mean_km[brand] = mean_km

brand_mean_km
```

```python
bmp_series = pd.Series(brand_mean_price)
pd.DataFrame(bmp_series, columns = ["mean_price"])
```

```python
bmkm_series = pd.Series(brand_mean_km)
brand_data = pd.DataFrame(bmkm_series, columns = ["mean_km"])

brand_data
```

```python
brand_data["mean_price"] = bmp_series

brand_data
```

All brands have a similar odometer reading, which does not appear to be the main driver of price
